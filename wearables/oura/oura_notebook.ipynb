```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
def list_csv_files_with_paths(data_dir):
    # Ensure the directory exists
    if not os.path.exists(data_dir):
        print(f"The directory {data_dir} does not exist.")
        return []

    # List all files in the directory
    files = os.listdir(data_dir)

    # Filter out only CSV files and get their full paths
    csv_files_with_paths = [os.path.join(data_dir, file) for file in files if file.endswith('.csv')]

    return csv_files_with_paths
```

```python
data_dir = '/Users/guhansundar/Documents/GuData/ObjectiveSubjectiveHealth/wearables/oura/oura_export_08172025'
csv_files = list_csv_files_with_paths(data_dir)

use_keys = ['dailysleep', 'dailyresilience', 'dailystress', 'dailyreadiness', 'sleep', 'dailyspo2', 'dailycardiovascularage', 'dailyactivity']
hr_key = ['heartrate']

df_dict = {}
for fn in csv_files:
    tdf = pd.read_csv(fn)
    key = fn.split('/')[-1].split('_')[0]
    if key in use_keys:
        df_dict[key] = tdf
    elif key in hr_key:
        hrdf = tdf

```

```python

# def combine_dataframes(df_dict):
# Initialize an empty list to store the processed dataframes
processed_dfs = []
for key, df in df_dict.items():
    df['day'] = pd.to_datetime(df['day'])
    numeric_cols = df.select_dtypes(include='number').columns
    df = df.groupby('day')[numeric_cols].agg('mean')
    df_renamed = df.rename(columns={col: f"{col}-{key}" for col in numeric_cols})
    processed_dfs.append(df_renamed)

combined_df = pd.concat(processed_dfs, axis=1)

```

```python
fig, axes = plt.subplots()
sns.scatterplot(x='day', y='average_heart_rate-sleep', data=combined_df)
plt.show()
```
